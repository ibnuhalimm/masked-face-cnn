```python
from keras.preprocessing import image as krimg
import matplotlib.image as mpimpg
import matplotlib.pyplot as plt
import glob
import os

image_dir = os.path.join(os.getcwd(), 'datasets/v4/00_plotting/face/')

images = []
images_paths = []
for img_path in glob.glob(image_dir + '/*.jpg'):
    images.append(mpimpg.imread(img_path))
    images_paths.append(img_path)
    
plt.figure(figsize=(15, 10))
columns = 5

for i, image in enumerate(images):
    img = krimg.load_img(images_paths[i], target_size=(1024, 1024))
    
    rows = int(len(images) / columns + 1)
    plt.subplot(rows, columns, i + 1)
    # plt.xlabel()
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.subplots_adjust(wspace=0, hspace=0)
```
